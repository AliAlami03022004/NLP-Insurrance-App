# EDA and Matplotlib Plot Recipes

This notebook mirrors the main grading-oriented visualizations as executable code cells.

```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

PROJECT_ROOT = Path.cwd().parents[0]
DATA_PATH = PROJECT_ROOT / 'data' / 'processed' / 'clean_reviews_phase3.csv'
df = pd.read_csv(DATA_PATH, low_memory=False)
train_df = df[df['type'].astype(str).eq('train')].copy()
sns.set_theme(style='whitegrid')
train_df.shape
```

## Star Distribution

```python
star_counts = train_df['note'].value_counts().sort_index()
fig, ax = plt.subplots(figsize=(8, 5))
sns.barplot(x=star_counts.index.astype(str), y=star_counts.values, color='#2b8cbe', ax=ax)
ax.set_title('Star Distribution (Train Split)')
ax.set_xlabel('Stars')
ax.set_ylabel('Number of Reviews')
plt.show()
```

## Top Insurers by Volume

```python
insurer_stats = (
    train_df.groupby('assureur', as_index=False)
    .agg(review_count=('note', 'size'), mean_star=('note', 'mean'))
    .sort_values('review_count', ascending=False)
    .head(15)
)
fig, ax = plt.subplots(figsize=(12, 6))
sns.barplot(data=insurer_stats, y='assureur', x='review_count', palette='Blues_r', ax=ax)
ax.set_title('Top 15 Insurers by Review Volume')
ax.set_xlabel('Review Count')
ax.set_ylabel('Insurer')
plt.show()
```

## Average Stars by Theme

```python
theme_stats = (
    train_df[train_df['note'].notna()]
    .groupby('theme_primary', as_index=False)
    .agg(avg_star=('note', 'mean'), reviews=('note', 'size'))
)
fig, ax = plt.subplots(figsize=(10, 5))
sns.barplot(data=theme_stats, x='theme_primary', y='avg_star', palette='crest', ax=ax)
ax.set_title('Average Stars by Theme')
ax.set_xlabel('Theme')
ax.set_ylabel('Average Stars')
ax.tick_params(axis='x', rotation=25)
plt.show()
```

## Review Length by Stars

```python
fig, ax = plt.subplots(figsize=(10, 5))
sns.boxplot(data=train_df, x='note', y='text_word_len', color='#99d8c9', showfliers=False, ax=ax)
ax.set_title('Review Length Distribution by Star Rating')
ax.set_xlabel('Stars')
ax.set_ylabel('Words per Review')
plt.show()
```
